```python
import uproot as up
import pandas as pd
import numpy as np
```

```python
hits = up.open("mlData.root:hits")
cells = up.open("mlData.root:cells")
particles = up.open("mlData.root:particles")
truth = up.open("mlData.root:truth")

print(hits.show())
print(cells.show())
print(particles.show())
print(truth.show())
```

```python
event = 0
lines = 11

print("hit_id,x,y,z,volume_id,layer_id,module_id")
for i in range(lines):
    print(hits["hit_id"].array()[event][i], hits["x"].array()[event][i], hits["y"].array()[event][i], hits["z"].array()[event][i], hits["volume_id"].array()[event][i], hits["layer_id"].array()[event][i], hits["module_id"].array()[event][i])

print()

print("hit_id,dep_charge,energyloss,volume_id,layer_id,module_id,sector_id,isochrone,skewed")
for i in range(lines):    
	print(cells["hit_id"].array()[event][i], cells["dep_charge"].array()[event][i], cells["energy_loss"].array()[event][i], cells["volume_id"].array()[event][i], cells["layer_id"].array()[event][i], cells["module_id"].array()[event][i], cells["sector_id"].array()[event][i], cells["isochrone"].array()[event][i], cells["skewed"].array()[event][i])
      
print()

print("hit_id, tx, ty, tz, tpx, tpy, tpz, particle_id")
for i in range(lines):
	print(truth["hit_id"].array()[event][i], truth["tx"].array()[event][i], truth["ty"].array()[event][i], truth["tz"].array()[event][i], truth["tpx"].array()[event][i], truth["tpy"].array()[event][i], truth["tpz"].array()[event][i], truth["particle_id"].array()[event][i])

print()

print("particle_id, vx, vy, vz, px, py, pz, nhits, pdgcode, start_time, primary")
for i in range(lines):
	print(particles["particle_id"].array()[event][i], particles["vx"].array()[event][i], particles["vy"].array()[event][i], particles["vz"].array()[event][i], particles["px"].array()[event][i], particles["py"].array()[event][i], particles["pz"].array()[event][i], particles["nhits"].array()[event][i], particles["pdgcode"].array()[event][i], particles["start_time"].array()[event][i], particles["primary"].array()[event][i])
```
